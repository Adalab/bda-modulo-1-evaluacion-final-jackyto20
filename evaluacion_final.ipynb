```python
import re

class TiendaOnline: 

   #creamos el metodo constructor con los atributos para agregar clientes y producto
   
    def __init__(self):

        self.inventario = [] #lista para guardar los productos que agreguemos al inventario
        self.clientes = {} #diccionario para guardar los datos de los clientes
        self.total_ventas = 0.0 #iniciamos las ventas en cero, ya que aún no hemos vendido


    #creamos método agregar producto al inventario
    def agregar_producto(self, nombre, precio, cantidad):

        #recorrer inventario para verificar si esta el producto en la lista
        for producto in self.inventario:
            if producto['nombre'] == nombre:
                producto['cantidad'] += cantidad #Si existe solo aumentamos su cantidad
                return f"producto {nombre} actualizado exitosamente. Nueva cantidad:{cantidad}"
           
        #Si el producto no esta en el inventario lo agregamos
        nuevo_producto = {'nombre': nombre, 'precio': precio, 'cantidad': cantidad}
        self.inventario.append(nuevo_producto)
        return f"Producto {nombre} ha sido agregado al inventario con cantidad {cantidad}."



    #Definir método para mostrar el inventario actualizado
    def ver_inventario(self):
        for nombre in self.inventario:
            return f"El inventario de productos es: {self.inventario}"
        

        
    #Definir el método para buscar producto por nombre
    def buscar_producto(self, nombre):
        #Recorrer el inventario para verificar si esta el producto
        for producto in self.inventario:
            if producto['nombre'] == nombre:
                return f"Nombre: {producto['nombre']}, Precio: ${producto['precio']}, Cantidad: {producto['cantidad']}"
        #Una vez recorre el inventario y no esa el producto que queremos buscar envia un mensaje informandolo       
        return f"El producto {nombre} no se encuentra en el inventario"            
            


    #Definir método actualizar stock (cantidad) del producto
    def actualizar_stock(self, nombre, cantidad):

        for producto in self.inventario:
            if producto['nombre'] == nombre:
                producto['cantidad'] +=cantidad
                return f"La cantidad del producto {nombre} ha sido actualizada. Este es el inventario actualizado {self.inventario}"
         
        return f"El producto {nombre} no esta en el inventario."



    #Definir método eliminar un producto del inventario
    def eliminar_producto(self, nombre):
        #recorrer el inventario para  buscar el producto
        for producto in self.inventario:
            #comparar el producto con el argumento que le envio
            if producto['nombre'] == nombre:
                #si coincide con el nombre se elimina
                self.inventario.remove(producto)
        #Devuelve mensaje indicando que el producto ha sido eliminado
        return f"El producto {nombre} ha sido eliminado. Este es el inventario actualizado {self.inventario}"
            
        
    
    
    #Definir metodo calcular_valor_inventario
    def calcular_valor_inventario(self):
        valor_total = 0 #inciamos el valor total en cero para que vaya aumentando
        for producto in self.inventario:
            valor_total += producto['precio'] * producto['cantidad'] # la variable valor total se va incrementando con el resultado de multiplicar precio x cantidad
        return f"El valor total del inventario es: $ {valor_total}"

    
    #Definir metodo buscar producto con regex
    def buscar_producto_regex(self, patron):
        
        productos_encontrados = []
        for producto in self.inventario:
            if re.search(patron, producto['nombre']):
                productos_encontrados.append(producto)

        if producto not in productos_encontrados:
                return f"no hay coincidencias con el patron"
              
        return  f"Los productos encontrados que inicia con 'cam' son :[{productos_encontrados}]"
        


    #Definir metodo para que el cliente realice compra
    def realizar_compra(self):
        carrito = []
        seguir_comprando = True
        while seguir_comprando == True:
            self.ver_inventario() #ver el inventario
            producto_comprar = input("Por favor ingrese el producto que desea comprar: ")
            cantidad_comprar = int(input(f"Por favor ingrese la cantidad de {producto_comprar} que quiere comprar: "))
            for producto in self.inventario:
                if producto['nombre'] == producto_comprar and cantidad_comprar <= producto['cantidad']:
                    compra = {'nombre': producto['nombre'], 'precio': producto['precio'], 'cantidad': cantidad_comprar}
                    carrito.append(compra)
                    producto['cantidad'] -= cantidad_comprar
                    #break
                else:
                    print("No hay suficiente cantidad disponible o el producto no está en el inventario.")

            seguir = int(input("Digite 1 para seguir comprando, 2 para terminar: "))
            if seguir != 1:
                seguir_comprando = False
                   
        return carrito
    

            
    #Definir el método para procesar el pago
    def procesar_pago(self, carrito):
                
        total_compra1 = 0 #iniciamos la compra en cero
        for producto in carrito:
            total_compra1 += producto['precio'] * producto['cantidad']
        
        #usamos incluimos el try indicar el codigo que puede generar error 
        try:
            pago = float(input(f"El total de este pedido es $ {total_compra1}. Ingrese el valor a pagar: "))
            if pago >= total_compra1:
                cambio = pago - total_compra1
                print(f"Gracias por su compra, el pago se ha realizado correctamente. el total es {total_compra1} Su cambio es ${cambio}.")
                self.total_ventas += total_compra1
            else:
                print("Monto insuficiente para completar la compra.")
        except ValueError: #en caso de que el usuario ingrese un valor no valido le mostrara este mensaje
            return f"Por favor, ingrese un número válido."

        


    # Definir método para agregar un cliente
    def agregar_cliente(self, nombre, email):
        if nombre not in self.clientes:
            self.clientes[nombre] = {'email': email, 'compras': []}
        else:
            print("El cliente ya está registrado.")

              
      

    # Definir método para ver los clientes registrados
    def ver_clientes(self):
        for nombre in self.clientes:
            return f"Esta es la lista actualizada de los clientes:{self.clientes}"
        
    
    
    # Defnir método para ver el historial de compras de un cliente
    def ver_compras_cliente(self, nombre_cliente):
        if nombre_cliente in self.clientes:
            print(f"Historial de compras de {nombre_cliente}:")
            for compra in self.clientes[nombre_cliente]['compras']:
                print(compra)
        else:
            print("El cliente no está registrado.")


        
    # Defnir método para registrar una compra en el historial del cliente
    def registrar_compra(self, nombre_cliente, carrito):
        total_compra= 0
        if nombre_cliente in self.clientes:
            self.clientes[nombre_cliente]['compras'].append(carrito)
            for producto in carrito:
                subtotal = producto['precio'] * producto['cantidad']
                total_compra += subtotal
            self.total_ventas += total_compra
            print(self.total_ventas)

        else:
            print("El cliente no está registrado.")


    #Defnir método par mostrar el historial de las compras del cliente
    def ver_compras_cliente(self, nombre_cliente):
        if nombre_cliente in self.clientes:
            print(f"Historial de compras de {nombre_cliente}:")
            for compra in self.clientes[nombre_cliente]['compras']:
                print(compra)
        else:
            print("El cliente no está registrado.")

        
 
    # Defenir método para calcular las ventas totales de la tienda
    def calcular_ventas_totales(self):
        print(f"Las ventas totales de la tienda son: ${self.total_ventas}")




mi_tienda = TiendaOnline()

# Agregamos nuevos productos al inventario
print(mi_tienda.agregar_producto('camisa', 20, 40))
print(mi_tienda.agregar_producto('blusas', 30, 30))
print(mi_tienda.agregar_producto('camiseta', 20, 50))
print(mi_tienda.agregar_producto('falda', 10, 39))
#Buscar un producto concreto en el inventario
print(mi_tienda.buscar_producto('camisa'))

# Ver el inventario actualizado
print(mi_tienda.ver_inventario())

#Actualizar el stock
print(mi_tienda.actualizar_stock('camisa', 1))

#Eliminar producto
print(mi_tienda.eliminar_producto('falda'))

#Calcular valor total del inventario
print(mi_tienda.calcular_valor_inventario())

#Buscar producto por regex
print(mi_tienda.buscar_producto_regex(r"cam*"))

# Realizar una compra
carrito_maria = mi_tienda.realizar_compra()

# Procesar el pago de esa compra
mi_tienda.procesar_pago(carrito_maria)

# Agregar un cliente
print(mi_tienda.agregar_cliente('Maria', 'maria@mail.com'))
print(mi_tienda.agregar_cliente('Pedro', 'pedro@gmail.com'))

#Ver clientes registrados
print(mi_tienda.ver_clientes())

# Registrar la compra en el historial del cliente
mi_tienda.registrar_compra('Maria', carrito_maria)

# Ver las compras de un cliente
mi_tienda.ver_compras_cliente('Maria')

# Ver ventas totales
mi_tienda.calcular_ventas_totales()









    
    
```
